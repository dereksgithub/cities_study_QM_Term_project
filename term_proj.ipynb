```python
city_df_selected = city_df[keep_list]
```

```python
# Drop rows with NaN values
citydf_cleaned = city_df_selected.dropna()

# Print the number of rows left
print(len(citydf_cleaned))
```

```python
citydf_cleaned 
```

Apply DBSCAN:

```python
from sklearn.cluster import DBSCAN
```

```python
# remove some string columns:
non_quant_cols_to_remove = ["City","Region", "City Type", "Metro-area", "Population Trend* "]

# removing some non-quantitative columns:
new_col_list = [x for x in keep_list if x not in non_quant_cols_to_remove]
len(new_col_list)
```

```python
# Select the columns for clustering
selected_columns = citydf_cleaned[new_col_list]

# Create an instance of DBSCAN
dbscan = DBSCAN(eps=0.5, min_samples=5)

# Fit the DBSCAN model to the selected columns
dbscan.fit(selected_columns)
```

```python
selected_columns
```

```python
# Assuming your DataFrame is named df

# Select all columns except the first one: city name.
selected_columns_dbscan = selected_columns.iloc[:, 0:]  
dbscan = DBSCAN(eps=0.1, min_samples=3)  # Adjust parameters as needed
dbscan.fit(selected_columns_dbscan)

labels = dbscan.labels_

selected_columns['DBSCAN_cluster'] = labels
```

```python
selected_columns
```

Apply K-means

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=5, random_state=0, n_init="auto").fit(selected_columns_dbscan)
kmeans.labels_
```

__Remark:__
K-means split the data points into 5 clusters.

```python
type(kmeans.labels_)
```

```python
selected_columns['KMeans_cluster'] = kmeans.labels_
```
